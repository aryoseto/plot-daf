# Plot Dynamic Amplification Factor

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Function to read and parse the input file
def to_list (filename) :
    a_list = []
    with open(filename, 'r') as toread :
        for line in toread :
            a_list.append([float(i) for i in line.split()])
    return a_list
```

```python
dataread = to_list('base-shear.txt')
```

```python
plt.rcParams["figure.figsize"] = (15,10)
plt.subplot(4,1,(1,3))
plt.plot([1/item[0] for item in dataread], [item[1]/item[2] for item in dataread])
plt.grid(True)

plt.subplot(4,1,4)
for item in dataread :
    a = [1/item[0], 1/item[0]]
    b = [0, 1]
    plt.plot(a, b, color ='black')
#plt.figure(figsize=(5, 5))
    
    
plt.show()
```
